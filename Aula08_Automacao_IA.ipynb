```python
# Define uma função para verificar a temperatura e exibir um alerta, se necessário.
def verificar_temperatura(temperatura_atual):
    # Verifica se a temperatura atual é maior que 30 graus Celsius.
    if temperatura_atual > 30:
        # Se for maior, imprime um alerta de aquecimento.
        print(f"ALERTA DE AQUECIMENTO: A temperatura atual " + str{temperatura_atual} + "°C o que recomenda o sistema")
    else:
        # Caso contrário, imprime uma mensagem indicando que a temperatura está normal.
        print(f"Temperatura normal: {temperatura_atual}°C.")

# --- Início da interação com o usuário ---
# Solicita ao usuário que insira a temperatura atual.
# A entrada é convertida para um número inteiro usando int().
try:
    temperatura_digitada = int(input("Por favor, digite a temperatura atual em °C: "))
    # Chama a função verificar_temperatura com o valor fornecido pelo usuário.
    verificar_temperatura(temperatura_digitada)
except ValueError:
    # Captura o erro se o usuário não digitar um número válido.
    print("Entrada inválida. Por favor, digite um número inteiro para a temperatura.")
# --- Fim da interação com o usuário ---
```
